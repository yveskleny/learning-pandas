```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_carros = pd.read_csv('car_prices.csv')
df_carros.head()
```

```python
df_carros.info()
df_carros = df_carros.dropna()
```

```python
df_carros.info()
```

```python
df_carros.columns
```

```python
maior_preco_venda_por_marca = df_carros.groupby('make')['sellingprice'].mean().sort_values(ascending=False)
maior_preco_venda_por_marca.head(10)
```

```python
maior_preco_venda_por_marca.head(5).plot(kind='bar')
plt.xlabel('Marca')
plt.ylabel('Preço Médio de Venda')
plt.title('Top 5 Marcas com Maior Preço Médio de Venda')
plt.show()
```

```python
df_carros.groupby('state').agg({
    'sellingprice': 'mean',
    'odometer': 'mean'
}).plot(kind='scatter', x='odometer', y='sellingprice')
plt.xlabel('Quilometragem Média')
plt.ylabel('Preço Médio de Venda')
plt.title('Relação entre Quilometragem e Preço de Venda')
plt.show()
```

```python
df_carros.groupby('year')['sellingprice'].mean().plot(kind='line', x='year', y='sellingprice')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço Médio de Venda')
plt.title('Variação do Preço Médio de Venda por Ano de Fabricação')
plt.plot()
```
