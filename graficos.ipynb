```python
import pandas as pd 
import matplotlib.pyplot as plt

dados_pessoas = pd.read_csv("dados/pessoas.csv")
dados_pessoas.head(10)
```

```python
dados_pessoas.groupby("cidade")["renda"].sum().plot(kind="line", title="Renda por Cidade")
plt.show()
```

```python
dados_produtos = pd.read_csv("dados/products-1000.csv")
dados_produtos.groupby("Category")["Price"].sum().sort_values(ascending=False).plot(kind="bar", title="Renda por Cidade")
plt.show()
```

```python
dados_pessoas.groupby("cidade")["compras_realizadas"].sum().plot(kind="pie", autopct='%1.1f%%', title="Distribuição das Compras")
plt.show()
```

```python
dados_pessoas["compras_realizadas"].plot(kind="hist", bins=5, title="Distribuição de Compras")
plt.show()
```

```python
dados_pessoas.plot(kind="scatter", x="renda", y="dias_na_plataforma", title="Renda x Dias")
plt.show()
```
