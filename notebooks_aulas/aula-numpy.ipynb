```python
import pandas as pd

df = pd.read_csv('../dados/vendas_produtos.csv')

df.head(10)

import random

def generate_random_hex_color():
    """Generates a random hex color code."""
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f"#{r:02x}{g:02x}{b:02x}"
```

```python
media_por_cidade = df.groupby('Cidade')["Lucro"].mean().sort_values(ascending=False)
std_por_cidade = df.groupby('Cidade')["Lucro"].std().sort_values(ascending=False)

lista_de_cores = [generate_random_hex_color() for _ in range(len(media_por_cidade))]
media_por_cidade.plot(kind='bar', yerr=std_por_cidade, capsize=10, color=lista_de_cores, edgecolor='black')
plt.ylabel('Lucro Mdio')
plt.title('Lucro Mdio por Cidade com Desvio Padro')
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
```
