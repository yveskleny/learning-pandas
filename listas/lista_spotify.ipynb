```python
import pandas as pd

path = "../dados/spotify_new.csv"

dados_spotify = pd.read_csv(path)

dados_spotify.head()
```

```python
dados_spotify.info()
```

```python
dados_spotify["listening_time"] = dados_spotify["listening_time"].fillna(dados_spotify.groupby("subscription_type")["listening_time"].transform("mean"))
dados_spotify

```

```python
dados_spotify["songs_played_per_day"] = dados_spotify["songs_played_per_day"].fillna(dados_spotify.groupby("subscription_type")["listening_time"].transform("mean"))
dados_spotify
```

### Manipulação

```python
dados_spotify.rename(columns={"listening_time": "minutos_por_dia"}, inplace=True)
dados_spotify.head(10)
```

```python
# dados_spotify.drop(columns=["ads_listened_per_week"], inplace=True)
```

```python
dados_spotify = dados_spotify[dados_spotify["age"] > 16]
dados_spotify
```

```python
dados_spotify["gender"] = dados_spotify["gender"].replace({"Male": "M", "Female": "F"})
dados_spotify.head(11)
```

```python
dados_spotify["country"].fillna("Desconhecido", inplace=True)
```
