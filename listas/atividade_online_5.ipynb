```python
import pandas as pd

path = "../dados/products-1000.csv"

df_produtos = pd.read_csv(path)

df_produtos.head()
```

```python
valores_filtrados = df_produtos[
    (df_produtos["Price"] > 500) &
    (df_produtos["Stock"] > 500) &
    (df_produtos["Size"] == "XL")
]
valores_filtrados
```

```python
df_produtos["Size"] = df_produtos["Size"].replace({"Small": "S", "Medium": "M", "Large": "L", "Extra Large": "XL"})
df_produtos.dropna(inplace=True)
df_produtos
```

```python
df_pessoas = pd.read_csv("../dados/pessoas.csv")

df_pessoas.head()
```

```python
import numpy as np

df_pessoas["gasto_medio"] = df_pessoas["renda"] / df_pessoas["compras_realizadas"]
df_pessoas["gasto_medio"] = df_pessoas["gasto_medio"].replace([np.inf, -np.inf], 0)
df_pessoas
```

```python
df_produtos = df_produtos.drop(labels=["Index", "Internal ID"], axis=1)
```

```python
df_produtos = df_produtos[
    (df_produtos["Size"] != "5x7 in") &
    (df_produtos["Size"] != '12x18 in')
]
df_produtos
```

```python
smartwatches_with_price_higher_than_700 = df_produtos[
    (df_produtos["Price"] > 700) &
    (df_produtos["Category"] == "Smartwatches")
]
smartwatches_with_price_higher_than_700
```

```python
low_stock = df_produtos[
    (df_produtos["Stock"] < 100) &
    (df_produtos["Availability"] == "limited_stock")
]
low_stock
```

```python
df_produtos = df_produtos.sort_values(by="Price", ascending=False)
```

```python
df_produtos = df_produtos.sort_values(by="Availability")
```

```python
df_produtos = df_produtos.sort_values(by="Stock")
```

```python
filtro = df_produtos[df_produtos["Category"] == "Automotive"]
filtro["Price"].mean()
```
