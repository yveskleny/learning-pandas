```python
import pandas as pd

df_imoveis = pd.read_csv("imoveis.csv")
df_imoveis
```

```python
df_imoveis["Vagas"] = df_imoveis["Vagas"].fillna(1)
df_imoveis
```

```python
imoveis_por_tipo_e_regiao = df_imoveis.drop(columns=["ID_Imovel", "Quartos", "Banheiros", "Vagas", "Área", "Status"])
imoveis_por_tipo_e_regiao
```

```python
imoveis_com_area_maior_que_100 = df_imoveis[df_imoveis["Área"] > 100]
imoveis_com_area_maior_que_100
```

```python
df_imoveis["Valor_por_m2"] = round(df_imoveis["Valor"] / df_imoveis["Área"], 2)
df_imoveis
```

```python
apartamentos_com_2_vagas_e_disponiveis = df_imoveis[(df_imoveis["Tipo"] == "Apartamento") & (df_imoveis["Vagas"] >= 2) & (df_imoveis["Status"] == "Disponível")]
apartamentos_com_2_vagas_e_disponiveis
```
