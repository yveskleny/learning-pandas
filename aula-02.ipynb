### Exercicio

```python
import pandas as pd
```

```python
dados_vendas = {
    "Produto": ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    "Preço Unitario": [1200, 25, 60, 350],
    "Quantidade Vendida": [10, 50, 30, 20]
}

df = pd.DataFrame(data=dados_vendas)
df.head()
```

```python
df["Valor Total"] = df["Preço Unitario"] * df["Quantidade Vendida"]
df.head()
```

### Atributos DataFrame

```python
# df.T -> para transpor matrizes
# df.at['linha', 'coluna'] -> utiliza os rotulos das colunas
# df.iat['indice_linha', 'indice_coluna'] -> utiliza os indices do dataframe começando do zero
# df.loc[['Rotulo A', 'Rotulo B']] -> posso pegar uma ou mais linhas passando o rotulo especifico ou uma lista de rotulos
# df.iloc[[1, 2]] -> posso pegar uma ou mais linhas passando o indice especifico ou uma lista de indice
df.iloc[:2,:2]
```

### Exercicio 2

```python


base_dados = {
    'Funcionario': ['Ana', 'Pedro', 'Maria', 'João', 'Sofia'],
    'Departamento': ['Vendas', 'TI', 'RH', 'Marketing', 'Vendas'],
    'Salario': [4500, 8000, 5500, 6000, 5000]
}
indices = ['ID-01', 'ID-02', 'ID-03', 'ID-04', 'ID-05']

df_empresa = pd.DataFrame(data = base_dados, index=indices)
df_empresa
```

```python
df_empresa.loc[['ID-03']]
```

```python
df_empresa.iloc[[1]]
```

```python
df_empresa.at['ID-04', 'Salario']
```

```python
df_empresa.loc[['ID-03']] = ['Maria Souza', 'RH', 6500]
df_empresa
```

```python
df_empresa.iloc[0:3, 2] += 200
df_empresa
```

```python
df_empresa.at['ID-02', 'Funcionario'] = 'Pedro Alves'
df_empresa
```

```python
df_empresa.iat[0, 1] = 'Marketing'
df_empresa
```
