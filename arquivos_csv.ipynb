```python
import json
produto = {
    "nome": "Teclado",
    "preco": 659.0,
    "quantidade": 29
}

def criar_json(dicionario):
    with open('produto.json', 'w', encoding='utf-8') as file:
        json.dump(dicionario, file, ensure_ascii=False, indent=4)



def ler_json(nome_arquivo):
    with open(nome_arquivo, 'r', encoding='utf-8') as file:
        novo_produto = json.load(file)
        print(f'Valor total: R$ {novo_produto['preco']*novo_produto['quantidade']:.2f}')
        return novo_produto
```

```python
criar_json(produto)
novo_produto = ler_json('produto.json')
```

```python

novo_produto['preco'] = 600

criar_json(novo_produto)

ler_json('produto.json')
```

```python
produtos = []
produtos.append({
    "nome": "Teclado",
    "preco": 659.0,
    "quantidade": 29
})
produtos.append({
    "nome": "Mouse",
    "preco": 289.0,
    "quantidade": 29
})

print(produtos)
criar_json(produtos)
```
